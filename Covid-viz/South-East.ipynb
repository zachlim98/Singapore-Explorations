```python
import pandas as pd
import plotly.express as plt
from datetime import date
```

```python
covid_data = pd.read_csv("owid-covid-data.csv")
```

```python
south_east_asia = ["Brunei","Cambodia","Indonesia","Laos","Malaysia","Myanmar","Philippines","Singapore","Thailand","Vietnam"]
sea_data = covid_data[(covid_data.location.isin(south_east_asia)) & (covid_data.date > "2021-1-1")]
```

```python
sea_data.to_csv("sea_data.csv")
```

```python
plt.scatter(sea_data, x="date", y="total_cases", facet_col="location", facet_col_wrap=2)
```
